import pandas as pd

class ProjectTracker:
    def __init__(self):
        self.df = pd.DataFrame(columns=["Task", "Owner", "Deadline", "Status"])

    def add_task(self, task, owner, deadline, status="Not Started"):
        # Count active tasks for this owner
        active_count = self.df[(self.df["Owner"] == owner) & (self.df["Status"].isin(["Not Started", "In Progress"]))].shape[0]

        if active_count >= 2 and status in ["Not Started", "In Progress"]:
            print(f"‚ùå Cannot assign task '{task}' to {owner}: already has 2 active tasks.")
            return

        new_task = {"Task": task, "Owner": owner, "Deadline": deadline, "Status": status}
        self.df = pd.concat([self.df, pd.DataFrame([new_task])], ignore_index=True)
        print(f"‚úÖ Task '{task}' added for {owner}.")

    def update_status(self, task, new_status):
        if task in self.df["Task"].values:
            self.df.loc[self.df["Task"] == task, "Status"] = new_status
            print(f"üîÑ Task '{task}' updated to {new_status}.")
        else:
            print(f"‚ö†Ô∏è Task '{task}' not found.")

    def view_tasks(self):
        print("\nüìã Project Tracker:")
        print(self.df.to_string(index=False))

import pandas as pd
from datetime import datetime, timedelta

# Sample data
data = [
    ("Alice", "2025-10-06", "2025-10-10"),
    ("Bob",   "2025-10-13", "2025-10-17"),
    ("Carol", "2025-10-20", "2025-10-24"),
    ("David", "2025-10-06", "2025-10-08"),
    ("Eva",   "2025-10-08", "2025-10-12"),
    ("Frank", "2025-10-20", "2025-10-24"),
    ("Grace", "2025-11-03", "2025-11-07"),
    ("Henry", "2025-11-03", "2025-11-05"),
]

df = pd.DataFrame(data, columns=["Employee", "Vacation Start", "Vacation End"])
df["Vacation Start"] = pd.to_datetime(df["Vacation Start"])
df["Vacation End"]   = pd.to_datetime(df["Vacation End"])

TOTAL_EMPLOYEES = 8
MAX_ON_VACATION = TOTAL_EMPLOYEES - 5  # = 3

def validate_schedule(vac_df):
    """Return True if no day has more than MAX_ON_VACATION people on vacation."""
    # build a date -> count map by expanding ranges
    counts = {}
    for _, row in vac_df.iterrows():
        start = row["Vacation Start"].date()
        end = row["Vacation End"].date()
        cur = start
        while cur <= end:
            counts[cur] = counts.get(cur, 0) + 1
            cur = cur + timedelta(days=1)
    # find any violation
    violations = {d:c for d,c in counts.items() if c > MAX_ON_VACATION}
    if violations:
        return False, violations
    return True, {}

def can_add_vacation(vac_df, employee, start_str, end_str):
    """Check if adding the new vacation would violate constraint."""
    new_row = {"Employee": employee,
               "Vacation Start": pd.to_datetime(start_str),
               "Vacation End": pd.to_datetime(end_str)}
    trial = pd.concat([vac_df, pd.DataFrame([new_row])], ignore_index=True)
    ok, violations = validate_schedule(trial)
    return ok, violations

def add_vacation(vac_df, employee, start_str, end_str):
    ok, violations = can_add_vacation(vac_df, employee, start_str, end_str)
    if not ok:
        print("‚ùå Cannot add vacation ‚Äî would violate availability constraint. Violations:")
        for d,c in violations.items():
            print(f"  {d}: {c} on vacation (max allowed = {MAX_ON_VACATION})")
        return vac_df
    new_row = {"Employee": employee,
               "Vacation Start": pd.to_datetime(start_str),
               "Vacation End": pd.to_datetime(end_str)}
    vac_df = pd.concat([vac_df, pd.DataFrame([new_row])], ignore_index=True)
    print(f"‚úÖ Vacation added for {employee}: {start_str} to {end_str}")
    return vac_df

# Example usage:

Saves a CSV vacation_tracker.csv for easy sharing.
